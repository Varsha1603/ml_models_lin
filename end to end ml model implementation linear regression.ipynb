```python
#independent and Dependent Features
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

#standardize the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Model Training

```python
from sklearn.linear_model import LinearRegression 
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
#print the coefficients and the intercepts
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
#on which parameters the model has been trained
regression.get_params()
```

```python
###prediction with test data
y_predict=regression.predict(X_test)
```

```python
y_predict
```

```python
##plot a scater plot for prediction


# Select one feature (e.g., first feature) for plotting
plt.scatter(X_test[:, 0], y_test, color='blue', label='Actual')   # Actual values
plt.scatter(X_test[:, 0], y_predict, color='red', label='Predicted') # Predicted values

plt.xlabel('Feature 1')  # Label for x-axis
plt.ylabel('Target')      # Label for y-axis
plt.legend()
plt.title('Actual vs Predicted Values')

```

```python
import pandas as pd
import seaborn as sns

# Convert test data and predictions to a DataFrame
df_g = pd.DataFrame(X_test, columns=cl_data.feature_names)
df_g['Actual'] = y_test
df_g['Predicted'] = y_predict

# Plot pair plot
sns.pairplot(df_g, vars=cl_data.feature_names[:], hue='Actual', palette='coolwarm')  # Show first 4 features

```

```python
fig, axs = plt.subplots(3, 3, figsize=(12, 8))
features = cl_data.feature_names[:]

for i, ax in enumerate(axs.flat):
    ax.scatter(X_test[:, i], y_test, color='blue', label='Actual', alpha=0.6)
    ax.scatter(X_test[:, i], y_predict, color='red', label='Predicted', alpha=0.6)
    ax.set_title(f"{features[i]} vs Target")
    ax.set_xlabel(features[i])
    ax.set_ylabel('Target')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print(mean_absolute_error(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
print(np.sqrt(mean_squared_error(y_test,y_predict)))
```

#mean_error should be less than base model mean

	MAE gives the average absolute difference between predicted and actual values  Easier to interpret.
	MSE penalizes larger errors more heavily  Useful when larger errors are more costly.
	A large gap between MSE and MAE suggests the presence of outliers or that large errors are skewing the model.


#performace matrix
Adjusted Square R Square 

adjusted R2 is always less than r2

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_predict)
score
```

```python
# adjusted r2 score
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
